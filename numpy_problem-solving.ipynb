```python
# Columns: [Age, Math Marks, Science Marks]
data = np.array([
    [18, 85, 78],   # Student 1
    [19, 92, 88],   # Student 2
    [17, 76, 95],   # Student 3
    [18, 65, 70],   # Student 4
    [20, 90, 85]    # Student 5
])
```

```python
#Get the shape of the matrix.
data.shape
```

```python
#Find the average age of students.
students = data[0:6,0:1]
print("no of the ages ",students)
average = np.mean(students)
print(average)
```

```python
#Extract Math marks of all students.
maths_marks = data[ : , 1:2]
maths_marks
```

```python
#Find the highest Science mark.
np.max(data[0:6, 2:])  
```

```python
#Get details of the student who scored more than 90 in Math.
data[data[:,1] > 90] 
```
