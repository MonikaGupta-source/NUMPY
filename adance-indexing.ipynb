**CONDITIONAL SELECTION**

```python
arr = np.array([10,20,30,40,50,60])
bool_num = print(arr[arr>20])
boolean_val = print(arr>20)
index_location = print(np.where(arr>20))#tells the index location based on the condition 
reserve_replacing = print(np.where(arr>20,arr,0))
#np.where(check-condition,reserve-variable,replace with)
```

**SORTING & ORDERING**

```python
arr = np.array([10,200,204,36,1,20,20,10,30,60])
sorted_array = print(np.sort(arr))
sort_indexes = print(np.argsort(arr))
```
